```python
import pandas as pd
```

```python
df = pd.read_csv('./decisionTree.csv')
```

```python
df = df.dropna()
```

```python
df
```

```python
inputs = df.drop('salary_more_then_100k',axis='columns')
```

```python
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

# ml work on number not in string so we need to change string into unwated number one way to use dummy data of pd another is using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

# now add the three column into inputs data

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
newInput = inputs.drop(['company','job','degree'],axis='columns')
```

```python
newInput
```

# import decision tree

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(newInput,target)
```

```python
model.score(newInput,target)
```

```python
model.predict([[2,2,1]])
```
